**Overview of the WineQuality Dataset**

Dataset Utilized: https://archive.ics.uci.edu/dataset/186/wine+quality

This dataset comprises of wine quality and its components. Each row represents a unique quality of wine, and each column contains numerical values for one of the unique quality found in wine, labeled from alcohol to flavanoids.

**Columns**

Product_Code: A unique identifier for each product.

Alcohol content to flavanoids: Unique qualities of wine


**Purpose**

The dataset is used to analyze wine quality and the unique contents it host over time. It serves as a primary tool for wine makers to understand the unique qualities that wines develop over time.


```python
from sklearn.datasets import load_wine
import pandas as pd

# Load the iris dataset from sklearn
wine = load_wine()

# Convert the iris dataset to a pandas dataframe
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# Add the target variable to the dataframe
#df['target'] = iris.target
df

```

```python
import pandas as pd

# Assuming df is your DataFrame
# Selecting the relevant columns
WQ = df[['alcohol', 'flavanoids']]

# Calculating the correlation matrix
corr_matrix = WQ.corr()

# Displaying the correlation coefficient between alcohol and flavanoids
correlation_coefficient = corr_matrix.loc['alcohol', 'flavanoids']
print(f"The correlation coefficient between alcohol and flavanoids is: {correlation_coefficient:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Assuming df is your DataFrame with the Alcohol and total_phenols
WQ = df.loc[:, 'alcohol':'flavanoids']

corr_matrix = WQ.corr()

plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # Mask for the upper triangle
heatmap = sns.heatmap(corr_matrix, mask=mask, annot=False, cmap='coolwarm', center=0, robust=True, linewidths=.5)
plt.title('Correlation Matrix of alcohol and total_Phenols', fontsize=16)
plt.show()
```

```python
import pandas as pd
from scipy.stats import pearsonr

# Assuming df is your DataFrame
# df = pd.read_csv('your_dataframe.csv')

# Selecting the relevant columns
WQ = df[['alcohol', 'flavanoids']]

# Calculating the correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(WQ['alcohol'], WQ['flavanoids'])

print(f"The correlation coefficient between alcohol and flavanoids is: {correlation_coefficient:.2f}")
print(f"The p-value is: {p_value:.4f}")

# Interpretation based on alpha = 0.05
if p_value < 0.05:
    print("The correlation is statistically significant at the 0.05 level.")
else:
    print("The correlation is not statistically significant at the 0.05 level.")
```
